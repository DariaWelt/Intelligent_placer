## Intelligent placer usage examplpe.
Code below is weak and did not contain exception handling because it is not functional part of tool. The only mission of this script is demonstration of lib working.

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from matplotlib import patches

from intelligent_placer_lib import check_image
```

```python
image_path = 'test/data/IMG_0.jpg'
image = np.array(Image.open('test/data/IMG_0.jpg'))
polygon = [[0, 0], [0.5, 0], [0.3, 0.4]]
polygon_mode = 'relative'
```

```python
plt.imshow(image)
plt.show()
```

```python
def draw_polygon(polygon, polygon_mode, image):
    fig, ax = plt.subplots()
    ax.imshow(image, cmap='gray')
    points = polygon.copy()
    if polygon_mode == 'relative':
        points = [[p[0] * image.shape[1], p[1] * image.shape[0]] for p in points]
    ax.add_patch(patches.Polygon(xy=points, fill=False, color='red'))

draw_polygon(polygon, polygon_mode, image)
```

```python
%%time
check_image(image_path, polygon, mode=polygon_mode, verbose=True)
```
